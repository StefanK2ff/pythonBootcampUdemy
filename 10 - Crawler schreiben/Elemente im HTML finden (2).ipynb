```python
rh = requests.get("https://www.lehrstelle-handwerk.de/ausbildung/lehrstellenboerse-praktikumsboerse/lehrstelle-suchen/list")

site = BeautifulSoup(rh.text, "html.parser")
```

```python
#print(site.prettify())
#counter = 0
for table in site.find_all("tbody"):
    beruf = table.select_one(".lehrstellen").text
    print(beruf.upper())
       
    for tr in table.find_all("tr"):
        
        #print(type(tr))
        
        if tr.select_one(".flush") == None:
            continue
        else: 
            print("=====")
            
            company = tr.select_one(".flush").text
            print(company)
            
            url = tr.select_one("a", href=True)
            print("https://www.lehrstelle-handwerk.de"+url['href'])
            
            start =  tr.select_one("a").text
            print(start)
            
            cident = str(url)[slice(86,str(url).find("?"))]
            print(cident)
            
            district = tr.select_one("td[data-label='Stadtteil']").text
            print(district)
            
            qualification = tr.select_one("td[data-label='Abschluss']").text
            
            print(qualification)
    print("=======================")
        
        
```
