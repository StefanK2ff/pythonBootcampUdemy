```python
xa = []
ya = []
xb = []
yb = []


search_a = ["Anna", "CA", "F"]
search_b = ["Bob", "CA", "M"]

with open("../data/names.csv", "r") as file:
    counter = 0
    for line in file:
        counter = counter + 1
        
        line_splitted = line.strip().split(",")
        if line_splitted[1] == search_a[0] and line_splitted[4] == search_a[1] and line_splitted[3] == search_a[2]: 
            xa.append(int(line_splitted[2]))
            ya.append(int(line_splitted[5]))
        elif line_splitted[1] == search_b[0] and line_splitted[4] == search_b[1] and line_splitted[3] == search_b[2]:
            xb.append(int(line_splitted[2]))
            yb.append(int(line_splitted[5]))

            
print(xa)
print(ya)
            
print(xb)
print(yb)
# 12345,Anna,1910,F,CA,999
# 32345,Anna,1911,F,CA,555
```
